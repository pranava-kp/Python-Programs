# Time series analysis : 
    A time series is a sequence of data points collected or recorded at successive equally spaced time intervals.
    
    Powerful tool for extracting insights, making predictions, and informing decision-making in a wide range of applications
    across industries. 
    
    It helps organizations and individuals leverage historical data to understand the past, anticipate the future.
    
    Helps in identifying trends, patterns, and seasonal variations in data over time.
    
    Allows for forecasting future values based on historical data.
    
    To monitor the performance of key metrics over time, such as sales, revenue, website traffic, and customer engagement.
    
    By analyzing time series data, organizations can identify areas for improvement and make data-driven decisions.
    
    Helps in detecting outliers in data that deviate from the expected patterns.
    

# Write a Python program to draw Time Series using Plotly Libraries.

```python
import pandas as pd
import plotly.express as px
tsla = pd.read_csv('Tesla.csv')
tsla.head()

#Volume : No.of units traded in a day
```

#P 2) Write a Python program for creating Maps using Plotly Libraries.
##### Plotly supports two different kinds of maps:
    Mapbox map():
        px.scatter_mapbox(), px.line_mapbox(), px.choropleth_mapbox() or px.density_mapbox() function
    Geo maps():
        px.scatter_geo(), px.line_geo() or px.choropleth() function

```python
import plotly.express as px
import pandas as pd

data=pd.read_csv("places.csv")
data.head(5)

```

```python
data.info()
```

```python
fig = px.scatter_mapbox(data, lat="LATITUDE", lon="LONGITUDE", hover_name="POIs")

fig.update_layout(mapbox_style="open-street-map")  #specify the style of the Mapbox map that we're creating

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})


fig.show()
```

```python
#px.scatter_geo
fig = px.scatter_geo(data, lat="LATITUDE", lon="LONGITUDE", hover_name="POIs")
fig.show()
```
