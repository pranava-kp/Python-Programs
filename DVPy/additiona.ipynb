```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Sample data for analysis (replace with your actual data)
data = {
'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
'Study Hours': [5, 7, 6, 4, 6],
'Test Score': [85, 90, 88, 80, 92],
'Attendance Percentage': [95, 97, 90, 92, 94]
}
# Convert data to a DataFrame
df = pd.DataFrame(data)
```

```python
highest_score_student = df.loc[df['Test Score'].idxmax()]['Student']
lowest_score_student = df.loc[df['Test Score'].idxmin()]['Student']
print("Student with the highest score:", highest_score_student)
print("Student with the lowest score:", lowest_score_student)
```

```python
correlation = df['Study Hours'].corr(df['Test Score'])
print("Correlation between study hours and test scores:", correlation)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Study Hours', bins=5, kde=True)
plt.title('Distribution of Study Hours')
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Student', y='Attendance Percentage', palette='pastel')
plt.title('Attendance Percentages Comparison')
plt.xlabel('Student')
plt.ylabel('Attendance Percentage')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=df, x='Student', y='Test Score', marker='o')
plt.title('Test Scores Change Among Students')
plt.xlabel('Student')
plt.ylabel('Test Score')
plt.xticks(rotation=45)
plt.show()
```

```python
import os

# Specify the directory
directory = r"C:\Users\prana\OneDrive\Desktop\New folder"

# Iterate over all files in the directory
for filename in os.listdir(directory):
    # Create the full file path
    old_file = os.path.join(directory, filename)
    
    # Check if it's a file (not a directory)
    if os.path.isfile(old_file):
        # Create the new file name
        new_file = os.path.join(directory, filename + ".jpg")
        
        # Rename the file
        os.rename(old_file, new_file)

print("File renaming completed.")
```
