PCA Sample Program

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Define your own dataset with 2 features
# Example data: Replace with your actual dataset
X = np.array([[2.5, 2.4],
              [0.5, 0.7],
              [2.2, 2.9],
              [1.9, 2.2]])
```

```python
# Apply PCA to reduce the data to 1 dimension
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)
```

```python
# Plot the original and PCA reduced data
plt.figure(figsize=(10, 5))
```

```python
# Original 2D data
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1])
plt.title('Original Data (2D)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
# 1D PCA reduced data
plt.subplot(1, 2, 2)
plt.scatter(X_pca, np.zeros_like(X_pca))
plt.title('PCA Reduced Data (1D)')
plt.xlabel('Principal Component 1')
plt.tight_layout()
plt.show()
```
