```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random data (100 normally distributed values)
np.random.seed(42)
# For reproducibility
data = np.random.randn(100)
# Create a figure with two subplots to compare different bin settings
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
# 1 row, 2 columns; figure size: 14x6 inches
# Histogram with 10 bins
axs[0].hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
axs[0].set_title("Histogram with 10 Bins")
axs[0].set_xlabel("Value")
axs[0].set_ylabel("Frequency")
# Histogram with 30 bins
axs[1].hist(data, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
axs[1].set_title("Histogram with 30 Bins")
axs[1].set_xlabel("Value")
axs[1].set_ylabel("Frequency")
# Adjust the layout to prevent overlapping
plt.tight_layout()
plt.show()
```
