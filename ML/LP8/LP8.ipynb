## Develop a program to demonstrate the working of the decision tree algorithm. Use Breast Cancer Data set for building the decision tree and apply this knowledge to classify a new sample.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
# Load the Breast Cancer dataset
data = load_breast_cancer()
```

```python
# Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
data.data, data.target, test_size=0.2, random_state=42)
```

```python
# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
```

```python
# Train the model using the training data
clf.fit(X_train, y_train)
```

```python
# Predict the labels for the test data and calculate the accuracy
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Predict the class for a new (first) test sample
sample = X_test[0]
predicted_label = clf.predict([sample])[0]
prediction_class = "Benign" if predicted_label == 1 else "Malignant"
print(f"Predicted Class for first test sample: {prediction_class}")
```

```python
# Visualize the decision tree structure
plt.figure(figsize=(30 , 10))
tree.plot_tree(clf,
filled=True,
feature_names=data.feature_names,
class_names=data.target_names)
plt.title("Decision Tree - Breast Cancer Dataset")
plt.show()
```
