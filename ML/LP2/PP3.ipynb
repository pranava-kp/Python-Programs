Analyze the relationship between different factors affecting student performance using a correlation matrix and a heatmap. The goal is to identify which factors have a strong positive or negative impact on a student's academic success.
The dataset consists of 100 students, each with the following attributes:
- Study Hours → Number of hours a student studies daily.
- Marks → Percentage score obtained in an exam.
- IQ → Intelligence Quotient (IQ score of the student).
- Sleep Hours → Number of hours the student sleeps daily.
- Stress Level → Self-reported stress level on a scale of 1 to 10.


Perform the below tasks
- Generate a synthetic dataset with realistic student performance attributes.
- Compute the correlation matrix to understand how study hours, IQ, sleep, and stress impact marks.
- Visualize the correlation matrix using a heatmap to identify strong positive/negative correlations.
- Interpret the results, explaining which factors significantly influence student performance.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generate a realistic dataset with some randomness
np.random.seed(42)
students = 100
data = {
'Study_Hours': np.random.randint(1, 10, students), # 1 to 10 hours
'Marks': np.random.randint(40, 100, students), # 40% to 100%
'IQ': np.random.randint(80, 140, students), # IQ between 80 and 140
'Sleep_Hours': np.random.randint(4, 10, students), # 4 to 10 hours of sleep
'Stress_Level': np.random.randint(1, 10, students) # Stress levels from 1 to 10
}
df = pd.DataFrame(data)
```

```python
# Compute correlation matrix
correlation_matrix = df.corr()
```

```python
# Plot heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Heatmap: Real-World Student Performance Correlations")
plt.show()
```
