Correlation analysis of California housing dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns # statistical data visualization library that provides functions for plotting including heatmaps.
import matplotlib.pyplot as plt # fundamental library for creating visualizations in Python.
from sklearn.datasets import fetch_california_housing
```

```python
# Load California housing dataset
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
```

```python
# calculates the Pearson correlation coefficients for all numerical column
corr_matrix = df.corr()
```

```python
# Plot heatmap
plt.figure(figsize=(10, 8))
```

```python
# Sets the figure size (10 inches wide, 8 inches tall).
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of California Housing Dataset")
plt.show()
```
